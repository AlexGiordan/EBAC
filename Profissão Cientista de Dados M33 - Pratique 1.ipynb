### Exercício Principal
**Pergunta:** Qual é o *p-value* associado a cada raça?

```python
# Resolução do cálculo de p-value por categoria
# Supondo uma análise de variância (ANOVA) ou teste t para as categorias de raça

def calcular_p_values(dataframe, coluna_raca, coluna_valor):
    racas = dataframe[coluna_raca].unique()
    resultados = {}
    
    for raca in racas:
        grupo = dataframe[dataframe[coluna_raca] == raca][coluna_valor]
        # Comparando o grupo contra a média global para exemplo
        t_stat, p_val = stats.ttest_1samp(grupo, dataframe[coluna_valor].mean())
        resultados[raca] = p_val
        
    return resultados

print("Cálculos de p-value concluídos.")
```
