```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregando a base de dados
# Certifique-se de que o arquivo .csv está no mesmo diretório
df = pd.read_csv('SINASC_RO_2019.csv')

# Visualizando as primeiras linhas para confirmar o carregamento
df[['IDADEMAE', 'PESO']].head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.rc('figure', figsize=(8, 6))
sns.boxplot(y='IDADEMAE', data=df)
plt.title('Distribuição da Idade da Mãe')
plt.ylabel('Idade da Mãe (anos)')
```

####  2) Faça um boxplot do peso do bebê

```python
plt.rc('figure', figsize=(8, 6))
sns.boxplot(y='PESO', data=df)
plt.title('Distribuição do Peso do Bebê')
plt.ylabel('Peso (gramas)')
```

####  3) Faça um histograma da idade da mãe

```python
plt.rc('figure', figsize=(10, 6))
sns.histplot(df['IDADEMAE'], bins=20, kde=True)
plt.title('Histograma: Idade da Mãe')
plt.xlabel('Idade da Mãe (anos)')
plt.ylabel('Frequência')
```

#### 4) Faça um histograma do peso do bebê

```python
plt.rc('figure', figsize=(10, 6))
sns.histplot(df['PESO'], bins=20, kde=True)
plt.title('Histograma: Peso do Bebê')
plt.xlabel('Peso (gramas)')
plt.ylabel('Frequência')
```
